```python
from torchvision import models
import torch
```

```python
vgg16_model = models.vgg16(pretrained = True) # Use ImageNet weights 
```

```python
from torchvision import transforms
preprocess = transforms.Compose([
    transforms.Resize((244,244)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.299,0.224, 0.225],
    )
])
```

```python
from PIL import Image
img = Image.open("bear.jpg")
img2=Image.open("dog2.jpg")
img3=Image.open("cat.jpg")
```

```python
img
```

```python
img2
```

```python
img3
```

```python
img=preprocess(img)
img2=preprocess(img2)
img3=preprocess(img3)
```

```python
type(img)
```

```python
img.shape
```

```python
# converting image into batch format
img=torch.unsqueeze(img,0)
img2=torch.unsqueeze(img2,0)
img3=torch.unsqueeze(img3,0)
```

```python
img.shape # added patch dimension at the first
```

```python
vgg16_model= vgg16_model.eval() # transform model from training to testing
```

```python
prob = vgg16_model(img)
prob2 = vgg16_model(img2)
prob3 = vgg16_model(img3)
```

```python
prob.shape
```

```python
value, index = torch.max(prob, 1)
value2, index2 = torch.max(prob2, 1)
value3, index3 = torch.max(prob3, 1)
```

```python
# to print index we must convert it from tensor to array
int(index.numpy()), int(index2.numpy()), int(index3.numpy())
```
